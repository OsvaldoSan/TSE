```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
data = load_iris()
data
```

```python
X = data['data']
X
```

```python
df = pd.DataFrame(X,columns=['x1','x2','x3','x4'])
df
```

```python
df.describe()
```

```python
from sklearn.manifold import MDS
```

```python
mds = MDS( n_components=2 )#, n_jobs=-1 )# n_jobs=-1 es para que trabaje la maquina con toda la potencia posible
type(mds)
```

```python
mds.fit_transform(df)
```

```python
Xm = pd.DataFrame(mds.fit_transform(df),columns=['d1','d2'])
Xm
```

```python
sns.lmplot(data=Xm,x='d1',y='d2',fit_reg=False)
```

```python
Xm['flor'] = data['target']
```

```python
sns.lmplot(data=Xm,x='d1',y='d2',fit_reg=False,hue='flor')
```
