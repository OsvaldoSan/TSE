```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

## Caso AND y OR

#### Generación de $S_n=\lbrace(x^{(i)},y^{(i)}),i=1,\ldots,n\rbrace$

```python
#Caso OR
n = 200
rs = np.random.RandomState(seed=1751)
x1 = ( rs.uniform(size=n) >=0.5 ).astype(int)
x2 = ( rs.uniform(size=n)>=0.5 ).astype(int)
```

```python
Sn = pd.DataFrame({'x1':x1,'x2':x2})
Sn['y'] = np.logical_and(Sn['x1'], Sn['x2']).astype(int)
Sn.head()
X = Sn[['x1','x2']]
X = X.values
y = Sn['y'] # Es una serie
y = y.values
Xt, Xv, yt, yv = train_test_split(X,y,train_size=0.7) # El 70 % de los datos ira a entrenamientoo
```

##### Verificación de que tan bueno es el modelo

```python
def accuracy_artesanal(X,y,theta):
    O=0
    tam=y.size
    for i in range(tam):
        O += ( np.dot(theta, X[i] ) == y[i] ).astype(int)
    return O/tam
```

```python
print(accuracy_artesanal(Xt,yt,theta))
print(accuracy_artesanal(Xv,yv,theta))
```

### Perceptron medio

```python
def perceptron_medio(y,X):
    T = 2
    n=y.size
    theta = np.array([0,0])
    thetag = np.array([0,0])
    c = 1
    for t in range(1,T):
        for i  in range(1,n):
            if y[i] * ( np.dot(theta, X[i] ) )<=0:
                theta += y[i]*X[i]
                thetag =  thetag + (y[i]*X[i])*c
            c = c - (1/( n*T ) )
            print("Theta=",theta)
            print("Theta gorro=",thetag)
    print(c)
    return thetag
```

### Implementación en Scikit-Learn

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
clasif = Perceptron(max_iter=T,fit_intercept=False)
```

```python
Sn = pd.DataFrame({'x1':x1,'x2':x2})
Sn['y'] = np.logical_or(Sn['x1'],Sn['x2']).astype(int)
Sn.head(2)
```

```python
X = Sn[['x1','x2']]
y = Sn['y'] # Es una serie

Xt, Xv, yt, yv = train_test_split(X,y,train_size=0.7) # El 70 % de los datos ira a entrenamiento

clasif.fit(Xt,yt)
```

##### $\vec{\theta}$

```python
clasif.coef_
```

##### $\theta_0$

```python
clasif.intercept_
```

##### Verificar que tan bueno es el modelo

```python
print( accuracy_score(y_true=yt , y_pred=clasif.predict(Xt) ) ) # Porcentaje de aciertos con el de entrenamiento
print(accuracy_score( y_true=yv, y_pred=clasif.predict(Xv) ) )# Porcentaje de aciertos con el de validación
```

```python
confusion_matrix(y_true=yv ,y_pred=clasif.predict(Xv) )# En Validación
```
